 Monte Carlo Simulation (Python)
Problem:
Simulate the price of a stock using the Geometric Brownian Motion (GBM) for 1000 simulations over 252 trading days.

Parameters:

Initial price: ₹100

Expected return (μ): 10%

Volatility (σ): 20%

```python
import numpy as np
import matplotlib.pyplot as plt

S0 = 100
mu = 0.10
sigma = 0.20
T = 1
N = 252
dt = T / N
simulations = 1000

np.random.seed(42)
price_paths = np.zeros((simulations, N))
price_paths[:, 0] = S0

for t in range(1, N):
    Z = np.random.standard_normal(simulations)
    price_paths[:, t] = price_paths[:, t - 1] * np.exp((mu - 0.5 * sigma ** 2) * dt + sigma * np.sqrt(dt) * Z)

# Plot
plt.figure(figsize=(10, 5))
for i in range(10):  # Plot only 10 paths
    plt.plot(price_paths[i])
plt.title("Simulated Stock Price Paths")
plt.xlabel("Days")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```
